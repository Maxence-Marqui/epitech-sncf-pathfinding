## Generate Datasets

```python
nbr = 1500
```

```python
import re
import random
import argparse

def read_all_lines(filename):
    file = open(filename, 'r', encoding='utf-8')
    lines = file.readlines()
    lines = [line.strip() for line in lines]
    file.close()
    return lines

sentences = read_all_lines("../generate_data/sentences")
towns = read_all_lines("../generate_data/towns")

trips = []
nom_ville = []

for nb in range(nbr):
    sentence = random.choice(sentences)
    town_start = random.choice(towns)
    town_end = town_start
    while town_end == town_start:
        town_end = random.choice(towns)

    trip = sentence
    trip = re.sub(r'\bSTART\b', town_start, trip)
    trip = re.sub(r'\bEND\b', town_end, trip)

    trips.append(trip)
    nom_ville.append(f'{town_start},{town_end}')
```

```python
data = {
    'PHRASE': trips,
    'NOM VILLE': nom_ville
}

gen_data = pd.DataFrame(data)
gen_data
```

## Datasets (jeu de données)

```python
data = pd.read_excel("jeu_données.xlsx")
data = data.drop(['Code erreur', 'Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8'], axis=1)
condition = (data['SYNTAXE'] == 1.0)
data = data[condition]
data = data.drop(['LANGUE', 'SYNTAXE'], axis=1)
data = data.reset_index(drop=True)
data
```

```python
data = gen_data
data
```

## Nettoyage des données

```python
nlp = spacy.load("fr_core_news_sm")

phrase_data = []

for i in range(0, len(data)):
    if "gare" in data["PHRASE"][i].lower():
        doc = nlp(data["PHRASE"][i].lower().replace("la gare de ", ""))
    else:
        doc = nlp(data["PHRASE"][i])

    verbes_a_supprimer = []
        
    for i, token in enumerate(doc):
        if token.pos_ == "ADP" and i > 0 and doc[i + 1].pos_ == "VERB":
            verbes_a_supprimer.append(i)
    
    doc = nlp(" ".join(token.text for i, token in enumerate(doc) if i not in verbes_a_supprimer))

    verbes_a_supprimer = []

    for i, token in enumerate(doc):
        if token.pos_ == "NOUN" and i > 0 and doc[i - 1].pos_ == "DET":
            verbes_a_supprimer.append(i)
    
    doc = nlp(" ".join(token.text for i, token in enumerate(doc) if i not in verbes_a_supprimer))
    
    phrases_filtres = ' '.join([token.text for token in doc if token.pos_ not in ['VERB', 'PRON', 'ADV', 'DET', 'AUX', 'CCONJ', 'SPACE']])
    
    phrases_filtres = phrases_filtres.replace('?', '').replace('!', '').replace('.', '').replace('pour', '')

    phrase_data.append(phrases_filtres)
```

```python
dataa = []
dataa = phrase_data
phrase_data = []
for k in range(0, len(dataa)):
    doc = nlp(dataa[k])
    phrases_filtres = ' '.join([token.text for token in doc if token.pos_ not in ['VERB', 'PRON', 'ADV', 'DET', 'AUX', 'CCONJ', 'SPACE']])
    phrases_filtres = phrases_filtres.replace(' - ', '-')
    phrase_data.append(phrases_filtres)

phrase_data
```

## Ajout des Flags "depart" et "arrivee" devant les localisations par rapport à la position

```python
datasets = []
train_data = []

found = 0

for i in range(0, len(phrase_data)):
    phrase = phrase_data[i].lower()

    liste_villes = data['NOM VILLE'][i].split(",")
    dep_arr = []
    for j in range(0, len(liste_villes)):
        dep_arr.append(liste_villes[j].lower())

    if len(dep_arr) == 2:
        dep = dep_arr[0].lower()
        arr = dep_arr[1].lower()
        depart_start = phrase.find(dep)
        depart_end = depart_start + len(dep)
        arrivee_start = phrase.find(arr)
        arrivee_end = arrivee_start + len(arr)

        #print((phrase.lower(), {"entities": [(depart_start, depart_end, "DEPART"), (arrivee_start, arrivee_end, "ARRIVEE")]}), dep, arr)

        #datasets.append((phrase.lower(), {"entities": [(depart_start, depart_end, "DEPART"), (arrivee_start, arrivee_end, "ARRIVEE")]}))
        datasets = (phrase.lower(), {"entities": [(depart_start, depart_end, "DEPART"), (arrivee_start, arrivee_end, "ARRIVEE")]})

        entites = datasets[1]['entities']
        
        index_present = any(-1 in entite for entite in entites)

        ## Quand entities ne contient pas le "-1"
        if (index_present == False) and (depart_start!=arrivee_start) and (depart_end!=arrivee_end):
            train_data.append(datasets)
            found +=1

train_data
```

## Train data - Spacy Model

```python
len(train_data)
```

```python
nlp = spacy.blank("fr")  # Charger un modèle de langue française vide

# Ajouter des entités personnalisées pour départ et arrivée
ner = nlp.add_pipe("ner")

# Ajouter les entités départ et arrivée au modèle
ner.add_label("DEPART")
ner.add_label("ARRIVEE")

# Données d'entraînement
TRAIN_DATA = train_data

# Initialiser l'entraînement du modèle
optimizer = nlp.begin_training()

# Boucle d'entraînement sur les données
# for i in range(len(TRAIN_DATA)):  # Modifier le nombre d'itérations selon les besoins
i = 0
for epoch in range(5):
    for text, annotations in TRAIN_DATA:
        print(i)
        doc = nlp.make_doc(text)
        example = Example.from_dict(doc, annotations)
        nlp.update([example], sgd=optimizer)
        i+=1
    print(f"Epoch {epoch + 1} completed.")

print("Training complete.")
```

## Prediction

```python
def result(txt):
    nlp = spacy.load("../model_2")
    detect_ville = []
    nl = spacy.load("fr_core_news_sm")
    test_text = txt.lower().replace("la gare de ", "").replace("gare de ", "").replace("la gare ", "").replace("gare ", "").replace("rendre", "")


    doc = nl(txt)
    localisation = []

    for ent in doc.ents:
        if ent.label_ == "LOC":  # Filtrez les entités qui sont des lieux
            localisation.append(ent.text.lower())

    ville = []
    print("LOC", localisation)
    for i in range(0, len(localisation)):
        ville.append(localisation[i].replace("la gare de ", "").replace("gare de ", "").replace("la gare ", "").replace("gare ", ""))

    doc = nlp(test_text.lower())

    if len(list(doc.ents)) == 2:
        for ent in doc.ents:
            for v in ville:
                if str(ent) in v:
                    detect_ville.append([ent.label_, ent.text])
#                     print("=>", ent.label_.lower(), ":", ent.text.upper())
    else:
        test_text = txt.lower().replace("la gare de ", "").replace("gare de ", "").replace("la gare ", "").replace("gare ", "").replace("rendre", "")
        doc = nlp(test_text.lower())

        for ent in doc.ents:
            for v in ville:
                if str(ent) in v:
                    detect_ville.append([ent.label_, ent.text])
#                     print("=>", ent.label_.lower(), ":", ent.text.upper())
                    
    return detect_ville
```

## Input Txt File => Output file

#### Detect language

```python
def detect_language(text):
    try:
        language = detect(text)
        return language
    except:
        return None
```

#### read input file

```python
res_lines = []

with open('sample_nlp_input.txt', 'r', encoding='utf-8') as file:
    for line in file:
        line = str(line.strip())
        elements = line.split(',')

        res = result(elements[1])

        if elements:
            number = elements[0].strip()
            if number.isdigit():
                if len(res) == 2:
                    if detect_language(str(elements[1])) == 'fr':
                        if res[0][0] == 'DEPART':
                            depart = res[0][1]
                            arrivee = res[1][1]
                        else:
                            depart = res[1][1]
                            arrivee = res[0][1]

                        print(number + "," + depart + "," + arrivee)
                        res_lines.append((number + "," + depart + "," + arrivee))
                    else:
                        print(number + ",", "NOT_FRENCH")
                        res_lines.append((number + ",", "NOT_FRENCH"))
                elif detect_language(str(elements[1])) != 'fr':
                    print(number + "," + "NOT_FRENCH" + ',')
                    res_lines.append((number + "," + "NOT_FRENCH" + ','))
                else:
                    print(number + "," + "NOT_TRIP" + ',')
                    res_lines.append((number + "," + "NOT_TRIP" + ','))
            else:
                print('0' + "," + "NOT_TRIP" + ',')
                res_lines.append(('0' + "," + "NOT_TRIP" + ','))
```

#### Store output file

```python
def storeSentence(text):
    with open('./nlp_output.txt', 'a', encoding='utf-8') as file:
        # Write content to the file
        file.write(f'{text}\n')
```

```python
for line in res_lines:
    storeSentence(line)
```

## Evaluation du modèle Spacy

```python
import spacy
from spacy.training.example import Example
from sklearn.metrics import accuracy_score, precision_score

# Load your spaCy model
nlp = spacy.load("../model_2")

# Evaluate the model
correct_predictions = 0
total_predictions = 0
true_positives = 0
false_positives = 0

for text, annotation in train_data:
    doc = nlp(text)
    
    # Extract true labels
    true_entities = set((start, end, label) for start, end, label in annotation.get("entities", []))
    
    # Extract predicted labels
    predicted_entities = set((ent.start_char, ent.end_char, ent.label_) for ent in doc.ents)
    
    # Calculate accuracy
    correct_predictions += len(true_entities.intersection(predicted_entities))
    total_predictions += len(true_entities)

    # Calculate true positives and false positives
    true_positives += len(true_entities.intersection(predicted_entities))
    false_positives += len(predicted_entities - true_entities)
    
# Calculate precision
precision = true_positives / (true_positives + false_positives) if (true_positives + false_positives) > 0 else 0
print(f"Precision: {precision * 100:.2f}%")

# Calculate accuracy
accuracy = correct_predictions / total_predictions
print(f"Accuracy: {accuracy * 100:.2f}%")
```
