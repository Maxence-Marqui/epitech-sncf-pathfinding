## Ajout des Flags "depart" et "arrivee" devant les localisations

```python
nlp = spacy.load("fr_core_news_sm")

depart = "<depart>"
arrivee = "<arrivee>"

nouvelle = []

for k in range(0, len(data)):
    phrase = data['PHRASE'][k]
    villes = data['NOM VILLE'][k]
    liste_villes = villes.split(",")
    dep_arr = []
    for j in range(0, len(liste_villes)):
        dep_arr.append(liste_villes[j].replace(' ', ''))

    doc = nlp(phrase)
    if len(dep_arr) == 2:
        for i in range(0, len(doc)):
            if doc[i].lower_==dep_arr[0].lower():
                nouvelle_phrase = phrase.replace(str(doc[i]), f"{depart}{doc[i]}")
            if doc[i].lower_==dep_arr[1].lower():
                nouvelle_phrase = nouvelle_phrase.replace(str(doc[i]), f"{arrivee}{doc[i]}")
                nouvelle.append(nouvelle_phrase)
```

```python
donnees = {
    'Phrases': nouvelle
}

df = pd.DataFrame(donnees)
df
```

## Ajout des Entities avec les Flags "depart" et "arrivee" (ca depend de la position des localisations dans la phrase)

#### Exemple

```python
phrase = "je voudrais aller de Toulouse à bordeaux"
depart_start = phrase.find("Toulouse")
depart_end = depart_start + len("Toulouse")
arrivee_start = phrase.find("bordeaux")
arrivee_end = arrivee_start + len("bordeaux")

resultat = (phrase, {"entities": [(depart_start, depart_end, "DEPART"), (arrivee_start, arrivee_end, "ARRIVEE")]})
print(resultat)
```

#### Basé sur Data

```python
datasets = []

for i in range(0, len(data)):
    phrase = data['PHRASE'][i].lower()

    liste_villes = data['NOM VILLE'][i].split(",")
    dep_arr = []
    for j in range(0, len(liste_villes)):
        dep_arr.append(liste_villes[j].replace(' ', ''))

    if len(dep_arr) == 2:
        dep = dep_arr[0].lower()
        arr = dep_arr[1].lower()
        depart_start = phrase.find(dep)
        depart_end = depart_start + len(dep)
        arrivee_start = phrase.find(arr)
        arrivee_end = arrivee_start + len(arr)        

        datasets.append((phrase.lower(), {"entities": [(depart_start, depart_end, "DEPART"), (arrivee_start, arrivee_end, "ARRIVEE")]}))

datasets
```

#### Datasets pour le Model

```python
dd = []
for k in range(0, 25):
    dd.append(datasets[k])
```

#### Model Spacy

```python
nlp = spacy.blank("fr")  # Charger un modèle de langue française vide

# Ajouter des entités personnalisées pour départ et arrivée
ner = nlp.add_pipe("ner")

# Ajouter les entités départ et arrivée au modèle
ner.add_label("DEPART")
ner.add_label("ARRIVEE")

# Données d'entraînement
TRAIN_DATA = dd

# Initialiser l'entraînement du modèle
optimizer = nlp.begin_training()

# Boucle d'entraînement sur les données
for i in range(20):  # Modifier le nombre d'itérations selon les besoins
    for text, annotations in TRAIN_DATA:
        doc = nlp.make_doc(text)
        example = Example.from_dict(doc, annotations)
        nlp.update([example], sgd=optimizer)
```

```python
# Test du modèle
test_text = "Je veux aller à lille depuis paris"
doc = nlp(test_text.lower())
for ent in doc.ents:
    print(ent.text, ent.label_)
```

```python
phrase_data = []

for i in range(0, len(data)):
    doc = nlp(data["PHRASE"][i])

    verbes_a_supprimer = []
    
    for i, token in enumerate(doc):
        if token.pos_ == "ADP" and i > 0 and doc[i + 1].pos_ == "VERB":
            verbes_a_supprimer.append(i)
    
    doc = nlp(" ".join(token.text for i, token in enumerate(doc) if i not in verbes_a_supprimer))

    verbes_a_supprimer = []

    for i, token in enumerate(doc):
        if token.pos_ == "NOUN" and i > 0 and doc[i - 1].pos_ == "DET":
            verbes_a_supprimer.append(i)
    
    doc = nlp(" ".join(token.text for i, token in enumerate(doc) if i not in verbes_a_supprimer))

    phrases_filtres = ' '.join([token.text for token in doc if token.pos_ not in ['VERB', 'PRON', 'ADV', 'DET', 'AUX', 'CCONJ']])

    phrase_data.append(phrases_filtres)
```

## Villes Departs + Arrivées

```python
nlp = spacy.load("fr_core_news_sm")  # Charger le modèle SpaCy en français

ville_trouvee = []
ville_non_trouvee = []

for i in range(0, len(phrase_data)):
    phrase = phrase_data[i]
    # Appliquer le modèle SpaCy à la phrase
    doc = nlp(phrase)
    
    trouver_arrivee = False  # Indicateur pour commencer la recherche après "vers"
    trouver_depart = False
    locations = []  # Liste pour stocker les villes
    locations_dep = []
    
    # Parcourir les tokens (mots) dans le texte
    for token in doc:
        for l in range(0, len(arrival)):
            if trouver_arrivee and token.ent_type_ == "LOC":  # Vérifier si l'entité est une localité
                locations.append(token.text)  # Ajouter la localité à la liste des locations
            elif token.text == arrival[l]:
                trouver_arrivee = True  # Activer la recherche des mots suivant "vers"

        for m in range(0, len(depart)):
            if trouver_depart and token.ent_type_ == "LOC":  # Vérifier si l'entité est une localité
                locations_dep.append(token.text)  # Ajouter la localité à la liste des locations
            elif token.text == depart[m]:
                trouver_depart = True

    # Imprimer la ville après "vers" (si elle existe)
    if len(locations) > 0:
        if len(locations) > 0 and len(locations_dep) > 0:
            print("++++++", doc)
            print(f"   ====> Depart : {locations_dep[0]}\n   ====> Arrivée: {locations[0]}")
        #ville_trouvee.append(locations[0])
    else:
        ville_non_trouvee.append(phrase)
ville_trouvee
```

## Villes Arrivées

```python
nlp = spacy.load("fr_core_news_sm")  # Charger le modèle SpaCy en français

ville_trouvee = []
ville_non_trouvee = []

for i in range(0, len(phrase_data)):
    phrase = phrase_data[i]
    # Appliquer le modèle SpaCy à la phrase
    doc = nlp(phrase)
    
    trouver_vers = False  # Indicateur pour commencer la recherche après "vers"
    locations = []  # Liste pour stocker les villes
    
    # Parcourir les tokens (mots) dans le texte
    for token in doc:
        for l in range(0, len(arrival)):
            if trouver_vers and token.ent_type_ == "LOC":  # Vérifier si l'entité est une localité
                locations.append(token.text)  # Ajouter la localité à la liste des locations
            elif token.text == arrival[l]:
                trouver_vers = True  # Activer la recherche des mots suivant "vers"
    
    # Imprimer la ville après "vers" (si elle existe)
    #print(doc)
    if len(locations) > 0:
        ville_trouvee.append(locations[0])
    else:
        ville_non_trouvee.append(phrase)
ville_trouvee
```
