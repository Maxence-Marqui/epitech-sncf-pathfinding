```python
import re
import random
import argparse

def read_all_lines(filename):
    file = open(filename, 'r', encoding='utf-8')
    lines = file.readlines()
    lines = [line.strip() for line in lines]
    file.close()
    return lines

sentences = read_all_lines("../generate_data/sentences")
towns = read_all_lines("../generate_data/towns")

trips = []
nom_ville = []

for nb in range(1000):
    sentence = random.choice(sentences)
    town_start = random.choice(towns)
    town_end = town_start
    while town_end == town_start:
        town_end = random.choice(towns)

    trip = sentence
    trip = re.sub(r'\bSTART\b', town_start, trip)
    trip = re.sub(r'\bEND\b', town_end, trip)

    trips.append(trip)
    nom_ville.append(f'{town_start},{town_end}')
```

```python
data = {
    'PHRASE': trips,
    'NOM VILLE': nom_ville
}

gen_data = pd.DataFrame(data)
data = gen_data
data
```

## Get depart et arrival words

```python
df = pd.read_excel('jeu_données.xlsx', sheet_name='Termes')
```

```python
arrival = df['Arrival words']
arrival = list(arrival)
arrival = [element.lower() for element in arrival if not (isinstance(element, float) and np.isnan(element))]
arrival = list(set(arrival))
```

```python
depart = df['Departure words']
depart = list(depart)
depart = [element.lower() for element in depart]
depart = list(set(depart))
```

## Detect Depart et Arrivee

#### Nettoyer les données [verb, pron, adv....] + condition pour supprimer "la gare de " des phrases

```python
nlp = spacy.load("fr_core_news_sm")

phrase_data = []

for i in range(0, len(data)):
    if "gare" in data["PHRASE"][i].lower():
        doc = nlp(data["PHRASE"][i].lower().replace("la gare de ", ""))
    else:
        doc = nlp(data["PHRASE"][i])

    verbes_a_supprimer = []
        
    for i, token in enumerate(doc):
        if token.pos_ == "ADP" and i > 0 and doc[i + 1].pos_ == "VERB":
            verbes_a_supprimer.append(i)
    
    doc = nlp(" ".join(token.text for i, token in enumerate(doc) if i not in verbes_a_supprimer))

    verbes_a_supprimer = []

    for i, token in enumerate(doc):
        if token.pos_ == "NOUN" and i > 0 and doc[i - 1].pos_ == "DET":
            verbes_a_supprimer.append(i)
    
    doc = nlp(" ".join(token.text for i, token in enumerate(doc) if i not in verbes_a_supprimer))
    
    phrases_filtres = ' '.join([token.text for token in doc if token.pos_ not in ['VERB', 'PRON', 'ADV', 'DET', 'AUX', 'CCONJ', 'SPACE']])
    
    phrases_filtres = phrases_filtres.replace('?', '').replace('!', '').replace('.', '').replace('pour', '')

    phrase_data.append(phrases_filtres)

phrase_data
```

## Villes Departs + Arrivées

```python
nlp = spacy.load("fr_core_news_sm")  # Charger le modèle SpaCy en français

ville_trouvee = []
ville_non_trouvee = []

for i in range(0, len(phrase_data)):
    phrase = phrase_data[i]
    # Appliquer le modèle SpaCy à la phrase
    doc = nlp(phrase)
    
    trouver_arrivee = False  # Indicateur pour commencer la recherche après "vers"
    trouver_depart = False
    locations = []  # Liste pour stocker les villes
    locations_dep = []
    
    # Parcourir les tokens (mots) dans le texte
    for token in doc:
        for l in range(0, len(arrival)):
            if trouver_arrivee and (token.ent_type_ == "LOC" or token.ent_type_ == "PER"):  # Vérifier si l'entité est une localité
                locations.append(token.text)  # Ajouter la localité à la liste des locations
            elif token.text == arrival[l]:
                trouver_arrivee = True  # Activer la recherche des mots suivant "vers"

        for m in range(0, len(depart)):
            if trouver_depart and (token.ent_type_ == "LOC" or token.ent_type_ == "PER"):  # Vérifier si l'entité est une localité
                locations_dep.append(token.text)  # Ajouter la localité à la liste des locations
            elif token.text == depart[m]:
                trouver_depart = True

    if len(locations) > 0 and len(locations_dep) > 0:
        res = data["NOM VILLE"][i].split(',')
        
        if ((locations_dep[0].lower() in res[0].lower()) and (locations[0].lower() in res[1].lower()) and locations_dep[0].lower() != "de" and locations[0].lower() != "de"):
            print("Phrase:", doc)            
            print(f"   ====> Depart : {locations_dep[0].upper()}\n   ====> Arrivée : {locations[0].upper()}")
            ville_trouvee.append((locations_dep[0], locations[0]))
        else:
            ville_non_trouvee.append((locations_dep[0], locations[0]))
```

```python
print("Total phrases:", len(phrase_data))
print("Total des résultats trouvés:", len(ville_trouvee))
print("Total des résultats non trouvés:", len(ville_non_trouvee))
```
